```python
import matplotlib.pyplot as plt
from astropy.io import fits


hdu1=fits.open('m53.fit')

datas1=hdu1[1].data
# data column
colorM53=datas1.field(4)
MagM53=datas1.field(3)



fig, ax = plt.subplots()
ax.invert_yaxis()
ax.scatter(colorM53,MagM53-16.27,8,c='k',alpha=1,label='M53')
ax.scatter(0.43,10.17,32,c='r',alpha=1,label='Star In Question 2')

fig.set_size_inches(20, 15)
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')

legend.get_frame().set_facecolor('#00FFCC')

plt.xlabel('B-V',fontsize=20)
plt.ylabel('Mv',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

#plt.plot(0.43,11.67,'ro')
plt.show()
```
