```python
import matplotlib.pyplot as plt
from astropy.io import fits



hdu1=fits.open('m53.fit')
hdu2=fits.open('m3.fit')

datas1=hdu1[1].data
datas2=hdu2[2].data
# data columns
col=datas2.columns
colorM53=datas1.field(4)
MagM53=datas1.field(3)
#m2
colorM3=datas2.field(4)
MagM3=datas2.field(2)
#m2

fig, ax = plt.subplots()
ax.invert_yaxis()
ax.scatter(colorM53,MagM53-16.27,15,c='k',alpha=1,label='M53')

ax.scatter(colorM3,MagM3-15.08,15,c='r',alpha=1,label='M3')
fig.set_size_inches(20, 15)
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')

legend.get_frame().set_facecolor('#00FFCC')



plt.xlabel('B-V',fontsize=20)
plt.ylabel('Mv',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
#plt.savefig('./M53M3.png')
```
