### 3.Introduction to Saha Equation
##### The Saha Equation

$$\cfrac{n_{r+1}n_{e}}{n_{r}}=\cfrac{G_{r+1}g_{e}}{G_{r}}\cfrac{(2\pi{m_{e}kT})^{3/2}}{h^3}exp({\cfrac{-\chi_{r}}{kT}})$$

##### Ionization Energy
$$E=-\cfrac{13.6Z^{2}eV}{n^{2}}$$

##### (a) For H inoization plasma

$$\cfrac{n_{e}^2}{n_{H}}=\cfrac{(2\pi{m_{e}kT})^{3/2}}{h^3}exp({\cfrac{-13.6eV}{kT}})$$
$$n_{e}+n_{H}=n_{total}$$

Where $n_{H}$ is hydrogen number density,$n_{e}$ is electron numbenr density,there is $n_{H^{+}}=n_{e}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

```

```python

T=np.arange(1000,100000,100)
ratio=np.arange(0,1,0.01)
n_total=10**16#total density of electro and h_plus, SI unit.
n_total2=10**18
n_total3=10**20
n_total4=10**22
n_total5=10**9

m_e=9.1*10**(-31)
k=1.38*10**(-23)
h=6.63*10**(-34)
c=(1/h**3)*(2*np.pi*m_e*k)**(3/2)#constant of saha equation in (a)
chi=13.6*1.6*10**(-19)#ionization potential in SI
T,ratio=np.meshgrid(T,ratio)
saha=(n_total*(ratio**2)/(1-ratio))-c*T**(3/2)*np.exp(-chi/(k*T))+1#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
saha2=(n_total2*(ratio**2)/(1-ratio))-c*T**(3/2)*np.exp(-chi/(k*T))+1#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
saha3=(n_total3*(ratio**2)/(1-ratio))-c*T**(3/2)*np.exp(-chi/(k*T))+1#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
saha4=(n_total4*(ratio**2)/(1-ratio))-c*T**(3/2)*np.exp(-chi/(k*T))+1#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
saha5=(n_total5*(ratio**2)/(1-ratio))-c*T**(3/2)*np.exp(-chi/(k*T))+1#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
```

```python
mpl.style.use('seaborn')
plt.figure(1)
fig=plt.figure(figsize=(30,20))
ax1=plt.subplot(111)
ax1.set_xlabel('$ T(K)$',fontsize=30)
ax1.set_ylabel('$({n_{H^{+}}}/{n_{total}})$',fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.title('H ionization fraction',fontsize=40)
plt.xlim(2000,20000)
#ax1.semilogx()
#ax1.semilogy()

#major_ticks = np.arange(1000, 100000, 100)                                              
#minor_ticks = np.arange(1000, 100000, 20)                                               
#ax1.set_xticks(major_ticks)                                                       
#ax1.set_xticks(minor_ticks, minor=True)                                           



#plot
plt.contour(T,ratio,saha5,1,colors='C4',linewidths=4,label='n=10^9/m^3')
plt.contour(T,ratio,saha,1,colors='C0',linewidths=4,label='n=10^16/m^3')
plt.contour(T,ratio,saha2,1,colors='C1',linewidths=4,label='n=10^18/m^3')
plt.contour(T,ratio,saha3,1,colors='C2',linewidths=4,label='n=10^20/m^3')
plt.contour(T,ratio,saha4,1,colors='C3',linewidths=4,label='n=10^22/m^3')
plt.text(2500,0.8,'n=10^9/m^3',fontsize=30)
plt.text(5000,0.8,'n=10^16/m^3',fontsize=30)
plt.text(7500,0.8,'n=10^18/m^3',fontsize=30)
plt.text(10000,0.8,'n=10^20/m^3',fontsize=30)
plt.text(17000,0.8,'n=10^22/m^3',fontsize=30)
plt.show()
#ax1.legend(fontsize=25)
```

##### (b) For He ionization

<font size=4>$$\cfrac{n_{r+1}n_{e}}{n_{r}}=\cfrac{2G_{r+1}}{G_{r}}\cfrac{(2\pi{m_{e}kT})^{3/2}}{h^3}exp({\cfrac{-\chi_{r}}{kT}})$$</font>

For first ionization of helium:${\chi}=24.48eV,2G_{r+1}/G_{r}=4$

For second ionization of helium:${\chi}=54.17eV,2G_{r+1}/G_{r}=1$

```python
########First Inonization of helium##########
T=np.arange(1000,100000,100)
ratio=np.arange(0,1,0.01)
n_total=10**16

chi1=24.48*1.6*10**(-19)
chi2=54.17*1.6*10**(-19)
T,ratio=np.meshgrid(T,ratio)
sahahelium=(n_total*(ratio**2)/(1-ratio))-4*c*T**(3/2)*np.exp(-chi1/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahahelium2=(n_total*(ratio**2)/(1-ratio))-1*c*T**(3/2)*np.exp(-chi2/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.


mpl.style.use('seaborn')
plt.figure(2)
fig=plt.figure(figsize=(30,20))
ax1=plt.subplot(111)
ax1.set_xlabel('$T(K)$',fontsize=30)
ax1.set_ylabel('$({n_{He^{+}}}/{n_{total}})$',fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.title('$He^{n+} ionization fraction$',fontsize=40)
plt.xlim(1000,35000)
#ax1.semilogx()
#ax1.semilogy()


plt.contour(T,ratio,sahahelium,[0],colors='C3',linewidths=4,label='n=10^16')
plt.contour(T,ratio,sahahelium2,[0],colors='C4',linewidths=4,label='n=10^16')
plt.text(10000,0.8,'$He^{+}$',fontsize=40)
plt.text(25000,0.8,'$He^{2+}$',fontsize=40)



plt.show()
```

##### (c) For C ionization

$$\cfrac{n_{r+1}n_{e}}{n_{r}}=\cfrac{2G_{r+1}}{G_{r}}\cfrac{(2\pi{m_{e}kT})^{3/2}}{h^3}exp({\cfrac{-\chi_{r}}{kT}})$$

For $C^{+}$,  ${\chi}=1086.5kJ/mol,2G_{r+1}/G_{r}=4/5$

For $C^{2+}$,  ${\chi}=2352.6kJ/mol,2G_{r+1}/G_{r}=1$

For $C^{3+}$,  ${\chi}=4620.5kJ/mol,2G_{r+1}/G_{r}=4$

For $C^{4+}$,  ${\chi}=6222.7kJ/mol,2G_{r+1}/G_{r}=1$

For $C^{5+}$,  ${\chi}=37831kJ/mol,2G_{r+1}/G_{r}=4$

For $C^{6+}$,  ${\chi}=47277kJ/mol,2G_{r+1}/G_{r}=1$

```python
T=np.arange(1000,1000000,100)
ratio=np.arange(0,1,0.01)
n_total=10**16
Na=6.0*10**23


chi1=1086.5/Na*1000
chi2=2352.6/Na*1000

chi3=4620.5/Na*1000
chi4=6222.7/Na*1000

chi5=37831/Na*1000
chi6=47277/Na*1000

T,ratio=np.meshgrid(T,ratio)
sahacarboni=(n_total*(ratio**2)/(1-ratio))-0.8*c*T**(3/2)*np.exp(-chi1/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahacarbon2i=(n_total*(ratio**2)/(1-ratio))-1*c*T**(3/2)*np.exp(-chi2/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahacarbon3i=(n_total*(ratio**2)/(1-ratio))-4*c*T**(3/2)*np.exp(-chi3/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahacarbon4i=(n_total*(ratio**2)/(1-ratio))-1*c*T**(3/2)*np.exp(-chi4/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahacarbon5i=(n_total*(ratio**2)/(1-ratio))-4*c*T**(3/2)*np.exp(-chi5/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.
sahacarbon6i=(n_total*(ratio**2)/(1-ratio))-1*c*T**(3/2)*np.exp(-chi6/(k*T))#rewriten saha equation by ratio,ratio=n_e/n_total,right side should be zero.






mpl.style.use('seaborn')
plt.figure(3)
fig=plt.figure(figsize=(30,20))
ax1=plt.subplot(111)
ax1.set_xlabel('$T(K)$',fontsize=30)
ax1.set_ylabel('$({n_{C^{n+}}}/{n_{total}})$',fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.title('$C{n^{+}} ionization fraction$\nn~10^16/m^3',fontsize=40)
plt.xlim(3000,1000000)
ax1.semilogx()
#ax1.semilogy()


#major_ticks = np.arange(3, 6, 1)                                              
#minor_ticks = np.arange(3, 6, 0.5)                                               
#ax1.set_xticks(major_ticks)                                                       
#ax1.set_xticks(minor_ticks, minor=True)    



plt.contour(T,ratio,sahacarboni,[0],colors='b',linewidths=12,label='n=10^16')
plt.contour(T,ratio,sahacarbon2i,[0],colors='k',linewidths=10,label='n=10^16')
plt.contour(T,ratio,sahacarbon3i,[0],colors='g',linewidths=8,label='n=10^16')
plt.contour(T,ratio,sahacarbon4i,[0],colors='purple',linewidths=6,label='n=10^16')
plt.contour(T,ratio,sahacarbon5i,[0],colors='yellow',linewidths=6,label='n=10^16')
plt.contour(T,ratio,sahacarbon6i,[0],colors='aqua',linewidths=6,label='n=10^16')


plt.text(4000,0.8,'CII',fontsize=40)
plt.text(10000,0.8,'CIII',fontsize=40)
plt.text(19000,0.8,'CIV',fontsize=40)
plt.text(30000,0.8,'CV',fontsize=40)
plt.text(120000,0.8,'CVI',fontsize=40)
plt.text(200000,0.8,'CVII',fontsize=40)



plt.show()
```
