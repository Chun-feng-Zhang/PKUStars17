```python
import matplotlib.pyplot as plt
from astropy.io import fits



hdu1=fits.open('m53.fit')


datas1=hdu1[1].data

# data columns
#col=datas2.columns
colorM53=datas1.field(4)
MagM53=datas1.field(3)
#m2
#colorM3=datas2.field(4)
#MagM3=datas2.field(2)
#m2

fig, ax = plt.subplots()
ax.invert_yaxis()
ax.scatter(colorM53,MagM53,15,c='k',alpha=1,label='M53')
fig.set_size_inches(20, 15)
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')

legend.get_frame().set_facecolor('#00FFCC')
#arrow
plt.annotate('HB', xy=(0, 17), xytext=(0, 16),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('Blue Straggler', xy=(0.2, 19), xytext=(-0.2, 20),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('Main Sequence Star', xy=(0.5, 21), xytext=(1, 21),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('RG', xy=(0.65, 18), xytext=(0.9, 18),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('RG', xy=(0.7, 17), xytext=(0.9, 18),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('RG', xy=(1.0, 15.3), xytext=(0.9, 18),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('AGB', xy=(0.68, 16.1), xytext=(0.5, 15),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('AGB', xy=(0.8, 15.9), xytext=(0.5, 15),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('AGB', xy=(0.85, 15.6), xytext=(0.5, 15),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.annotate('RR', xy=(0.4, 16.8), xytext=(0.3, 16),
            arrowprops=dict(facecolor='red', shrink=0.05),fontsize=20,color='red')

plt.xlabel('B-V',fontsize=20)
plt.ylabel('Vmag',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
#plt.savefig('./M53_6TYPES.png')
```
